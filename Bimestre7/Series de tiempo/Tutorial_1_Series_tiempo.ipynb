# Importar Librerias

```python
import pandas as pd
from darts import TimeSeries
import matplotlib.pyplot as plt
from scipy.stats import boxcox
import numpy as np
from darts.datasets import AirPassengersDataset
```

# Cargar data de libreria darts

```python

dataset = AirPassengersDataset()
series = dataset.load()
```

```python
series.plot()
plt.show()
```

# Transformaci√≥n de Serie Temporal

## Logaritmo

```python
log_data = np.log(series.pd_series())
log_series = TimeSeries.from_series(log_data)

log_series.plot()
plt.show()
```

## Diferenciaci√≥n

```python
diff_series = series.diff()
diff_series.plot()
plt.show()
```

# Box-Cox

```python
values = series.pd_series().values
boxcox_values, lambda_value = boxcox(values)
print(f'Lambda: {lambda_value}')
```

```python
boxcox_series = TimeSeries.from_values(boxcox_values)
boxcox_series.plot()
plt.show()
```
