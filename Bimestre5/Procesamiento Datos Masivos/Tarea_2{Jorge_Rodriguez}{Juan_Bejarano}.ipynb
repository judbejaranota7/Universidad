### Pregunta 2

2.  Examina el código fuente de la página web:

El método para hacerlo depende del navegador que estés utilizando.

Considera que esta página contiene un catálogo de libros, el catálogo cuenta con un total de 1000 libros, pero se muestran 20 a la vez, por lo que para revisarlos todos debes hacer repetidamente click en el botón “next”.



```python
# Lista Vacía para agregar la data de las 50 páginas
all_data = []


while url:

    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    book_titles = soup.find_all('h3')

    for title in book_titles:
        all_data.append(title.text.strip())

    # Encontrar la próxima página
    next_button = soup.find('li', class_="next")
    if next_button:
        next_page_url = next_button.find('a')['href']
        url = url.rsplit('/', 1)[0] + '/' + next_page_url
    else:
        url = None

# Print de los nombres de los libros en las 50 páginas.
for data in all_data:
    print(data)
```

### Pregunta 3

3.  Escribe un programa en python que imprima en la consola:

Los links de todos los libros que tengan una calificación de 5 estrellas en la página 1 de
http://books.toscrape.com/
.



```python
# Obtener los libros con calificación de 5 estrellas en la página 1

url = "http://books.toscrape.com/"
page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")

books = soup.find_all("article", class_="product_pod")
for book in books:
    star_rating = book.find("p", class_="star-rating")["class"][-1]
    if star_rating == "Five":
        book_title = book.h3.a["title"]
        book_link = url + book.h3.a["href"]

        print(f"{book_title}, {book_link}")
```

### Pregunta 4

4.  Extiende tu programa para extraer e imprimir:

Todos los libros del catálogo completo (las 50 páginas) cuyo rating sea igual a 5 estrellas.

```python
url = "http://books.toscrape.com/"
page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")

# Lista Vacía
five_star_books = []

# Obtenemos la info de las 50 páginas
while url:
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Encuentra los libros con 5 estrellas de la página actual
    books = soup.find_all("article", class_="product_pod")
    for book in books:
        star_rating = book.find("p", class_="star-rating")["class"][-1]
        if star_rating == "Five":
            book_title = book.h3.a["title"]
            book_url = url + book.h3.a["href"]
            five_star_books.append((book_title, book_url))

    # Encuentra la próxima página
    next_button = soup.find('li', class_="next")
    if next_button:
        next_page_url = next_button.find('a')['href']
        url = url.rsplit('/', 1)[0] + '/' + next_page_url
    else:
        url = None

# Print de los nombres, URL de los libros
for book in five_star_books:
    print(f"{book[0]}, {book[1]}")
```
